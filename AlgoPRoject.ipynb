```python
from PIL import Image
from numpy import*
import matplotlib.pyplot as plt
import numpy as np

name='d6ddddd4cec20d1daaa145353c290ff4--cute-love-drawings-simple-doodles-drawings.jpg'
temp=Image.open(name)
temp=temp.convert('1')      # Convert to black&white
A = array(temp)             # Creates an array, white pixels==True and black pixels==False
new_A=empty((A.shape[0],A.shape[1]),None)    #New array with same size as A
x = []
y = []
points=list()
for i in range(len(A)):
    for j in range(len(A[i])):
        if A[i][j]==True:
            new_A[i][j]=0
        else:
            point = list()
            new_A[i][j]=1
            x.append(i)
            y.append(j)
            point.append(i)
            point.append(j)
            points.append(point)

plt.figure(figsize=(150, 150))
mx = max(x)+1
my = max(y)+1
print min(x), max(x)+1,min(y), max(y)+1
plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.yticks(np.arange(min(y), max(y)+1, 1.0))
plt.scatter(x, y)

plt.show()
```

```python
def voronoi(points, shape=(600,600)):
    depthmap = np.ones(shape,numpy.float)*1e308
    colormap = np.zeros(shape,numpy.int)

    
    def hypot(X,Y):
        return (X-x)**2 + (Y-y)**2

    for i,(x,y) in enumerate(points):
        paraboloid = np.fromfunction(hypot,shape)
        colormap = np.where(paraboloid < depthmap,i+1,colormap)
        depthmap = np.where(paraboloid < depthmap,paraboloid,depthmap)
    print colormap
    return colormap

def convert_vornoi(points, colormap):
    color = colormap
    rows = len(colormap)
    cols = len(colormap[0])
    for i in range(0,rows):
        for j in range(0,cols):
            if j ==0 or j == cols-1:
                color[i][j] = 0
            elif ((colormap[i][j] != colormap[i][j+1])):
                color[i][j] = 0
            else:
                color[i][j] = 7
    for (x,y) in points:
        colormap[x,y] = 0
    print color
    return color

def draw_map(colormap):
    shape = colormap.shape

    palette = numpy.array([
            0x000000FF, #black
            0xFF0000FF, #red
            0x00FF00FF, #green
            0xFFFF00FF, #yellow
            0x0000FFFF, #blue
            0xFF00FFFF, #pink
            0x00FFFFFF, #turcoise
            0xFFFFFFFF, #white
            ])

    colormap = np.transpose(colormap)
    pixels = np.empty(colormap.shape+(4,),numpy.int8)

    pixels[:,:,3] = palette[colormap] & 0xFF
    pixels[:,:,2] = (palette[colormap]>>8) & 0xFF
    pixels[:,:,1] = (palette[colormap]>>16) & 0xFF
    pixels[:,:,0] = (palette[colormap]>>24) & 0xFF

    image = Image.frombytes("RGBA",shape,pixels)
    image.save('voronoi.png')
    print 'done'                
    
if __name__ == '__main__':
    #points = ([100,100],[350,300],[400,65],[324,145],[200,399])
    vor = voronoi(points, shape =(mx,my))
    vor = convert_vornoi(points,vor)
    draw_map(vor)
```
